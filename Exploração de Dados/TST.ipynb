```python
import pandas as pd
```

```python
df_hotel = pd.read_csv('http://harve.com.br/praticas/reservashotel.csv')
df_hotel.head(10)
```

```python
df_hotel.dtypes
```

```python
#Qual é a média de estadia de pessoas que tem bebê?

pessoas_com_bebe = df_hotel[df_hotel['bebês'] > 0] #filtrar pessoas com bebê
pessoas_com_bebe['estadia_total'] = (pessoas_com_bebe['estadias_em_noites_finalsemana'] + pessoas_com_bebe['estadias_em_noites_semana']) 
media_estadia_com_bebe = pessoas_com_bebe['estadia_total'].mean()
media_estadia_com_bebe
```
